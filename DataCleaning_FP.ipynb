```python
import pandas as pd
df = pd.read_csv('games.csv')
```

```python
df['Release date']
```

```python
df['Release date'] = pd.to_datetime(df['Release date'], format='%b %d, %Y', errors='coerce')
```

```python
df['Release_date_day'] = df['Release date'].dt.day
df['Release_date_month'] = df['Release date'].dt.month
df['Release date_year'] = df['Release date'].dt.year
```

```python
filter = df['Release date'].dt.year>=2018
```

```python
df = df[filter]
```

```python
df = df.drop(columns=['Screenshots','Movies'],axis=0)
```

```python
df = df.drop(columns='Release date')
```

```python
check = ['Website','Support url','Support email']
df[check]
```

```python
df = df.drop(columns=check,axis=0)
```

```python
df = df.drop(columns='Header image',axis=0)
```

```python
check = ['Metacritic url','Metacritic score']
df = df.drop(columns=check,axis=0)
```

```python
df = df.drop(columns='Notes',axis=0)
```

```python
df.isna().sum()
```

```python
check = ['Score rank','Reviews']
df = df.drop(columns=check,axis=0)
```

```python
df['Recommendations'].value_counts()
```

```python
check = ['Recommendations','Tags']
df = df.drop(columns=check,axis=0)
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['About the game'],axis=0)
```

```python
check = ['Developers','Publishers','Categories','Genres']
df[check]
```
